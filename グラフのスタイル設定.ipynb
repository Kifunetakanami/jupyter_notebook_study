```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')
import numpy as np
```

```python
# 色の設定

fig, ax = plt.subplots()

# 線の色を名前で指定
ax.plot([1, 3], [3, 1], label = 'aqua', color = 'aqua')

# 16進数のRGB値で指定
ax.plot([1, 3], [1, 3], label = '#0000FF', color = '#0000FF')

# RGBAをfloatで指定
ax.plot([1, 3], [2, 2], label = '(0.1, 0.2, 0.5, 0.3)', color = (0.1, 0.2, 0.5, 0.3))

# 凡例を表示
ax.legend()

plt.show()
```

```python
# 何故か動かない・・・・・

fig, ax = plt.subplots()

# 塗りつぶしの色を指定
ax.bar = ([1], [3], color = 'red')

# 塗りつぶしの色と枠線の色を指定
ax.bar = ([2], [4], color = 'red', edgecolor = 'black')

plt.show()
```

```python
fig, ax = plt.subplots()

# 5.5ポイントの幅の線で描画
ax.plot([1, 3], [3, 1], linewidth = 5.5, label = '5.5')

# 10ポイントの幅の線で描画
ax.plot([1, 3], [1, 3], linewidth = 10, label = '10')

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

# 破線で描画
ax.plot([1, 3], [3, 1], linestyle = '--', label = 'dashed')

# 一点鎖線で描画
ax.plot([1, 3], [1, 3], linestyle = '-.', label = 'dashdot')

# 点線で描画
ax.plot([1, 3], [2, 2], linestyle = ':', label = 'dotted')

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('xlabel', family = 'fantasy', size = 20, weight = 'bold')
ax.set_ylabel('ylabel', family = 'serif', size = 40, weight = 'light')
ax.set_title('graph title', family = 'monospace', size = 25, weight = 'heavy')

plt.show()
```

```python
# フォントのスタイルを辞書で定義
fontdict = {
    'family':'fantasy',
    'size':20,
    'weight':'normal',
}

fig, ax = plt.subplots()

# 辞書形式でフォントのスタイルを指定
ax.set_xlabel('xlabel', fontdict = fontdict)
ax.set_ylabel('ylabel', fontdict = fontdict)

# 個別指定したsizeで上書き可能
ax.set_title('graph title', fontdict = fontdict, size = 40)

plt.show()
```

```python
fig, ax = plt.subplots()

# Kifuneというテキストを描画
ax.text(0.2, 0.4, 'Kifune', size = 20)

plt.show()
```

```python
# 折れ線グラフ

import pandas as pd
import matplotlib.style
import matplotlib.pyplot as plt

# スタイルを指定
matplotlib.style.use('ggplot')

# DataFrameを作成
df = pd.DataFrame({'A':[1, 2, 3], 'B':[3, 1, 2]})

# 折れ線グラフを描画
df.plot()

plt.show()
```

```python
# 棒グラフ

import numpy as np

# ランダムに3行2列のデータを生成してDataFrameを作成
np.random.seed(123)
df2 = pd.DataFrame(np.random.rand(3, 2),
                   columns = ['y1', 'y2'])

# 棒グラフを描画
df2.plot.bar()

plt.show()
```

```python
# 積み上げ棒グラフを描画

df2.plot.bar(stacked = True)
plt.show()
```
