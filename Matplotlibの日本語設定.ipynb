```python
# Matplotlibの日本語設定

import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')
import numpy as np

fig, ax = plt.subplots()

ax.set_title('日本語タイトルを設定')
ax.text(2, 6, '文字列を描画', size = 20)

x = [1, 2, 3]
y = [10, 2, 3]

labels = ['スパム', 'ハム', '卵']
ax.bar(x, y, tick_label = labels, label = '食べ物')
ax.legend()

plt.show()
```
