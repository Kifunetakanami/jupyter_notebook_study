```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.style

# ggplotスタイルを指定
matplotlib.style.use('ggplot')
```

```python
# MATLABスタイル

# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y)            # 折れ線グラフを描画
plt.title('MATLAB-style') # グラフにタイトルを設定する

plt.show()                # グラフを表示
```

```python
# オブジェクト指向スタイル

# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

ax.plot(x, y)             # 折れ線グラフを描画
ax.set_title('OOP-style') # タイトルを設定

plt.show()                # グラフを表示
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2)      # 2つのサブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(2, 2) # 2行2列のサブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(ncols = 2) # 1行2列のサブプロットを配置
plt.show()
```

```python
import matplotlib.style

# スタイルの一覧を表示
print(matplotlib.style.available)
```

```python
# グラフのスタイルにclassicを指定
matplotlib.style.use('classic')

fig, ax = plt.subplots()
ax.plot([1, 2])

plt.show()
```

```python
fig, axes = plt.subplots(ncols = 2)

# サブプロットにタイトルを設定
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')

# 描画オブジェクトにタイトルを設定
fig.suptitle('figure title')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('x label') # X軸にラベルを設定
ax.set_ylabel('y label') # Y軸にラベルを設定

plt.show()
```

```python
fig, ax = plt.subplots()

# 凡例用のラベルを設定
ax.plot([1, 2, 3], [2, 4, 9], label = 'legend label')
ax.legend(loc = 'best') # 凡例を表示

plt.show()
```

```python
fig, ax = plt.subplots()

# 凡例用のラベルを設定
ax.plot([1, 2, 3], [2, 4, 9], label = 'legend label')
ax.legend(loc = 'lower right') # 凡例を右下に表示

plt.show()
```
